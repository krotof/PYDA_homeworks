```python
data['Prev'] = data['ClientUUId'].shift()
data['PrevDate'] = data['SaleDate'].shift()
```

```python
def weekrange(row):
    if(row['ClientUUId'] == row['Prev']):
        date1 = row['SaleDate'].split('.')
        date2 = row['PrevDate'].split('.')
        date1_obj = datetime.datetime.strptime(date1[0], '%Y-%m-%d %H:%M:%S')
        date2_obj = datetime.datetime.strptime(date2[0], '%Y-%m-%d %H:%M:%S')
        weeks_delta = date1_obj - date2_obj
        week_range = math.floor(weeks_delta.days / 7)
        return week_range
    else:
        return pd.NaT
```

```python
data['WeekRange'] = data.apply(weekrange, axis=1)
```

```python
data['SaleDate'] = pd.to_datetime(data['SaleDate'])
data['PrevDate'] = pd.to_datetime(data['PrevDate'])
data.dtypes
```

```python
data.drop(data.columns[[0,1]], axis=1, inplace=True)
data.info()
```

```python
def dayrange(row):
    if(row['ClientUUId'] == row['Prev']):
        date1 = row['SaleDate']
        date2 = row['PrevDate']
        days_delta = date1 - date2
        days_range = math.floor(days_delta.days)
        return days_range
    else:
        return pd.NaT
```

```python
data['DayRange'] = data.apply(dayrange, axis=1)
```

```python
filtered_data = data.filter(items = ['Rank', 'ClientUUId', 'DayRange'])
```

```python
group_results = filtered_data.groupby('ClientUUId').agg({'DayRange': 'mean', 'Rank': 'count'})
group_results.head()
```

```python
group_results.head(50)
```

```python
def avweeks(row):
    return math.floor(row['DayRange'] / 7)
```

```python
group_results['AverageWeeks'] = group_results.apply(avweeks, axis=1)
```

```python
group_results.rename({'DayRange': 'AverageDays', 'Rank': 'OrdersCount'}, axis=1, inplace=True)
```

```python
joined_result = data.join(group_results, on='ClientUUId')
```
